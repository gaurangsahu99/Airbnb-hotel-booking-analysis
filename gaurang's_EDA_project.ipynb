```python
import numpy as np                             #importing libraries
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from google.colab import drive                  #loading dataset
drive.mount('/content/drive')
```

```python
file_path="/content/drive/MyDrive/Airbnb NYC 2019.csv"
df=pd.read_csv(file_path)
df.head(2)                                      #viewing dataset
df.info()                                       #dataset info
df.axes[0]                                      #counting rows and columns
df.axes[1]
df.head()
```

```python
df1=df.fillna(0)                                 #replacing all none values with 0
df1.head()
```

```python
df1.head()
plt_1=df1["neighbourhood_group"].value_counts()              #extracting column
plt.plot(plt_1)
plt.title("region v/s no of properties")                     #labelling of the graph
plt.xlabel("neighbourhood_group")
plt.ylabel("number of properties")

```

we have used  a line chart to show frequecies of demand of hotels in particular areas in new york.


insights: as we can see from the graph demand is all time high in manhattan region of new york from which we can come up to a conclusion of various types of amenties in this area such as follws:
.
1.a famous tourist spot.
2.it may be a business hotspot area.




```python
plt_2= df1.groupby(['neighbourhood_group','room_type'], as_index=False)[['price']].mean()          #extracting columns using groupby operations
sns.barplot(data=plt_2, x='neighbourhood_group', y ='price',hue="room_type")                       #plotting using seaborn library.


```

we have used a barplot to show price distribution of different room types in different regions of new york. cost of living is very high in these areas.


INSIGHTS-----as we can see price of all different types of rooms in manhattan is highest from this we can conclude,it must be a posh area of new york,follwoed by brooklyn.

```python
sns.countplot(data=df1, x="neighbourhood_group", hue="room_type")               #seaborn countplot to valuecount roomtypes in different region.       

```

we have used a barplot to show different types of room types in different cities.

insights: we can conclude that people prefer private rooms over other room types except in manhattan where people rent entire apartment. 

```python
labels = df['room_type'].value_counts().index
sizes = df['room_type'].value_counts().values
plt.pie(sizes, labels=labels,autopct="%2.2f%%")
```

here,we have used a pie chart to show percentage distribution of different room types.

insights:people generally do not prefer shared room.people like privacy.

```python
sns.lineplot(data=df1, x="neighbourhood_group", y="reviews_per_month", hue="room_type")           #deployment of lineplot of seaborn library.
```

here we have used seaborn lineplot to show number of checkin in different regions of new york in different room types.

insights:manhattan people generally prefer entire room don't genrally reviews them on the other hand other people like to brag about they are in manhattan.

```python
plt_3=df1.groupby(["neighbourhood_group"])[["availability_365"]].mean()          
ax=plt_3.plot.bar(figsize=(10,5),fontsize=14)                                   #providing dimensions   
ax.set_title("region v/s availability(in days)",fontsize=20)
ax.set_xlabel("region",fontsize=20)
ax.set_ylabel("availability(in days)",fontsize=20)
```

we have used a barplot to show availability of rooms in different new york.

insights:brooklyn is the most busiest place in new york or most people in brooklyn are not from brooklyn.


```python
plt_4=df1.groupby(["neighbourhood_group"])[["number_of_reviews"]].mean()
plt.plot(plt_4)
plt.title("region v/s number of reviews")
plt.xlabel("region")
plt.ylabel("number of reviews")

```

we have used a line plot to plot number of reviews against regions of new york

insights: though people stays less in staten island but are more active in reviwes while people in manhattan are more kind of non active in reviewing properties.

```python
sns.pairplot(df1,vars=["price","number_of_reviews","reviews_per_month","availability_365"],hue="neighbourhood_group")    #using pairplot as mentioned in sample EDA
plt.show()
```

we have used a pair plot to show correlation between price,number of reviews,reviews per month and availability in different regions.

```python
corr_df=df1.corr()                                                  #using corr() command to get correlation between differnt values of dataset.
plt.figure(figsize=(10,5))
sns.heatmap(corr_df,annot=True)
```

we have used a heatmap to show correlation between different values of dataset.

insights:number of reviews and reviews per month has highest positive correlation.
